# Distribution of average daily entries/exits per MTA station

The original questions behind my analysis were: 
- At what MTA stations should composting organizations conduct outreach to encourage more individuals to compost?

- Where should a composting organization open a new drop-off site and, based on this answer, conduct outreach to promote it? 

To answer the first question, I wanted to look at the distribution of average daily entries/exits per MTA station.

The below histogram shows that the vast majority of stations had fewer than 5,000 entries/exits on average each day between April 24, 2021 and July 30, 2021. This histogram is right-skewed.

```python
# look at histogram of avg_daily_entries_exits
plt.hist(station_daily_avg_counts.avg_daily_entries_exits)
```

The box-and-whisker plot below shows that there are many stations that are outliers for having very high average daily entries/exits between April 24, 2021 and July 30, 2021.

```python
sns.boxplot(x='avg_daily_entries_exits', data=station_daily_avg_counts)
```

# Top, Middle, and Bottom MTA Stations by Avg. Daily Entries/Exits 

The below bar charts show 3 groupings of 5 stations each: 
   - those with the fewest average daily entries/exits between April 24, 2021 - July 30, 2021
   - those with the median average daily entries/exits between April 24, 2021 - July 30, 2021
   - those with the most average daily entries/exits between April 24, 2021 - July 30, 2021
   
Based on the box-and-whisker plot above, we know that the 5 stations with the most average daily entries/exits are outliers.

```python
stations_1 = list(low_daily_avg_counts.station)
stations_1_values = list(low_daily_avg_counts.avg_daily_entries_exits)

stations_2 = list(mid_daily_avg_counts.station)
stations_2_values = list(mid_daily_avg_counts.avg_daily_entries_exits)

stations_3 = list(high_daily_avg_counts.station)
stations_3_values = list(high_daily_avg_counts.avg_daily_entries_exits)

plt.figure(figsize=(16, 8))

plt.subplot(3, 1, 1)
plt.barh(stations_1, stations_1_values)
plt.title('MTA stations with the Fewest Avg. Daily Entries/Exits \n from late April 2021 - July 2021')

plt.subplot(3, 1, 2)
plt.barh(stations_2, stations_2_values)
plt.title('MTA stations with the Median Avg. Daily Entries/Exits \n from late April 2021 - July 2021')

plt.subplot(3, 1, 3)
plt.barh(stations_3, stations_3_values)
plt.tight_layout()
plt.title('MTA stations with the Most Avg. Daily Entries/Exits \n from late April 2021 - July 2021')
```

Based on the above, I think it would be best to conduct outreach to promote composting at the 5 stations with the median number of average daily entries/exits. While the volume of people is significantly higher at the 5 stations with the most average daily entries/exits - thus potentially increasing the chances of reaching more people - I also suspect that the high volume and crowdedness translates into people being less likely to want to stop and talk about composting. 

At the middle 5 stations, they might be more likely to do so. Also, approximately 3000 people entering and exiting those stations on average daily is still a decent sized sample of people to potentially reach.
